```python



#@authors  - abhimanyu,roshan,abhijeet
# =============================================================
#This code aggregates the tables folder to generate the files 
#only with rows that has exponential/logistic coefficients value against them and ultimately aggregate them 
#into one consolidated file which then can be used to feed into ML algorithms
#
#==================================================================================

import os
from glob import glob
from collections import defaultdict
import pandas as pd
import csv

path = "/Users/abhimanyuabhinav/Downloads/table_names_for_patterns"
EXT = "*.csv"
outpath = "/Users/abhimanyuabhinav/Downloads/table_names_for_patterns/files_folder/"

for file in glob(os.path.join(path,EXT)):
    print(file)
    file_name = file.split('/')[-1]
    file_in_focus = pd.read_csv(file, encoding = "ISO-8859-1")
    try:
        pattern_rows = file_in_focus[(file_in_focus["Exp_fall"]=='yes') | (file_in_focus["Exp_nonfit"]=='yes')|(file_in_focus["Exp_2date"]=='yes')| (file_in_focus["Exp_lin"]=='yes')|(file_in_focus["Exp_used"]=='yes')]
        pattern_rows.to_csv (outpath+file_name, index = True, header=True)
    except KeyError:
        print(f"{file} has no Exp_fall column")
    except UnicodeDecodeError:
        print(f"{file} has encoding error")
        
        
    
    

    
    
    
    
    
    
```
